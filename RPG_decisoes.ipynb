```python
import random

class NPC:
    def __init__(self, name, hp, attack_power):
        self.name = name
        self.hp = hp
        self.max_hp = hp  # Guardar o HP máximo para calcular percentuais
        self.attack_power = attack_power

    def decide_action(self, player_hp):
        # Definir probabilidades baseadas no estado de saúde do NPC
        if self.hp > 0.6 * self.max_hp:  # NPC com mais de 60% de vida é mais ofensivo
            actions = ["atacar", "defender", "curar"]
            probabilities = [0.6, 0.2, 0.2]  # Mais chance de atacar
        elif self.hp > 0.3 * self.max_hp:  # NPC com 30%-60% de vida é equilibrado
            actions = ["atacar", "defender", "curar"]
            probabilities = [0.4, 0.4, 0.2]  # Equilibrado entre atacar e defender
        else:  # NPC com menos de 30% de vida é mais defensivo e pode tentar fugir
            actions = ["atacar", "defender", "curar", "fugir"]
            probabilities = [0.2, 0.5, 0.2, 0.1]  # Mais chance de defender, pequena chance de fugir

        action = random.choices(actions, probabilities)[0]
        return action

    def perform_action(self, action):
        if action == "atacar":
            return random.randint(5, self.attack_power)
        elif action == "curar":
            heal_amount = 10
            self.hp = min(self.hp + heal_amount, self.max_hp)  # Curar sem passar do HP máximo
            return 0
        elif action == "defender":
            return -1  # -1 significa que o NPC está se defendendo
        elif action == "fugir":
            print(f"{self.name} tentou fugir do combate!")
            return None

def player_action():
    # Jogador escolhe a ação
    print("\nEscolha sua ação:")
    print("1. Atacar")
    print("2. Curar")
    print("3. Defender")
    choice = input("Digite o número da ação: ")

    if choice == "1":
        return "atacar"
    elif choice == "2":
        return "curar"
    elif choice == "3":
        return "defender"
    else:
        print("Escolha inválida, atacando por padrão.")
        return "atacar"

# Exemplo de combate com decisões interativas e probabilidades
npc = NPC("Guerreiro", 50, 15)
player_hp = 40
player_defending = False

while npc.hp > 0 and player_hp > 0:
    # Turno do NPC
    action = npc.decide_action(player_hp)
    print(f"\nO NPC decidiu: {action}")

    if action == "fugir":
        print("O NPC fugiu com sucesso!")
        break
    elif action == "atacar":
        damage = npc.perform_action(action)
        if player_defending:
            damage = max(0, damage - 5)  # Reduz o dano se o jogador estiver defendendo
        player_hp -= damage
        print(f"O NPC atacou causando {damage} de dano. HP do jogador: {player_hp}")
    elif action == "curar":
        npc.perform_action(action)
        print(f"O NPC se curou. HP do NPC: {npc.hp}")
    elif action == "defender":
        print("O NPC está se defendendo.")

    # Turno do jogador
    player_defending = False
    action = player_action()

    if action == "atacar":
        player_attack = random.randint(5, 10)
        npc.hp -= player_attack
        print(f"Você atacou causando {player_attack} de dano. HP do NPC: {npc.hp}")
    elif action == "curar":
        player_hp += 10
        print(f"Você se curou. Seu HP: {player_hp}")
    elif action == "defender":
        player_defending = True
        print("Você está se defendendo.")

if npc.hp <= 0:
    print("\nO NPC foi derrotado.")
if player_hp <= 0:
    print("\nVocê foi derrotado.")
```
